```python
import numpy as np
import scipy as sp
import scipy.special as spec
import matplotlib.pyplot as plt

t = np.arange(0, 10, .1)
n = 4
s = []

for j in t:
    s.append(spec.riccati_jn(n,j)[0])
plt.plot(t, s)

        
plt.show()
```

```python
def legendre_p(n, m, z):
    return spec.lpmv(n, m, z)

def legendre_tau(n, m, z):
    """ Returns generalized Legendre function tau

    Derivative is calculated based on relation 14.10.5:
    http://dlmf.nist.gov/14.10
    """
    return ((m-n-1) * legendre_p(n-1, m, z) - n * z * legendre_p(n, m, z))  / (1 - z*z)
ss1 = []
ss2 = []
for n in range(0,7):
    s1 = []
    t = np.arange(-1, 1, .01)
    for j in t:
        s1.append(legendre_p(n, n, j))
    ss1.append(s1)

    s2=[]
    for j in t:
        s2.append(legendre_tau(n, n, j))
    ss2.append(s2)

for s1 in ss1:
    plt.plot(t, s1)
plt.show()

for s2 in ss2:
    plt.plot(t, s2)
plt.show()


```
