```python
import numpy as np
import scipy as sp
import scipy.special as spec
import matplotlib.pyplot as plt

t = np.arange(0, 10, .1)
n = 4
s = []

for j in t:
    s.append(spec.riccati_jn(n,j)[0])
plt.plot(t, s)

        
plt.show()
```

```python
s = []

for j in t:
    s.append(spec.lpmn(n,n,j)[0][0])
plt.plot(t, s)
plt.show()
```

```python
def setup_legendres(m,n,x):
    t = np.arange(0,x,.1)
    pmn = dpmn = []
    for i in t:
        lpmn = spec.lpmn(m,n,i)
        pmn.append(lpmn[0])
        dpmn.append(lpmn[1])
    return(pmn, dpmn)

(pmn, dpmn) = setup_legendres(0,0,1.5)
print(pmn)
```
